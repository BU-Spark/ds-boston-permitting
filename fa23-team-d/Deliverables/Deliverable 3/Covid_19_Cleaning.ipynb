```python
import pandas as pd
import numpy as np
```

```python
covid = pd.read_csv("C://Users//郝治寰//Desktop//CS506 Final Project//Checkpoint A//Covid.csv")
```

```python
covid
```

```python
covid = covid.dropna()
```

```python
covid = covid[["Week End Date", "Confirmed deaths", "Confirmed cases"]]
```

```python
covid['Week End Date'] = pd.to_datetime(covid['Week End Date'])
covid_cleaned = covid.sort_values(by='Week End Date')
```

```python
covid_cleaned
```

```python

# Create a copy of the original DataFrame to work with
covid_year_month = covid_cleaned.copy()

# Transform 'date' column into 'year_month' format for the first DataFrame
covid_year_month['year_month'] = covid_year_month['Week End Date'].dt.to_period('M')

# Aggregate confirmed cases and confirmed deaths by year and month
covid_year_month = covid_year_month.groupby('year_month').agg({'Confirmed cases': 'sum', 'Confirmed deaths': 'sum'}).reset_index()

# Create a separate DataFrame to calculate yearly sum of cases and deaths
covid_yearly = covid_cleaned.copy()

# Extract year from 'date' column for the second DataFrame
covid_yearly['year'] = covid_yearly['Week End Date'].dt.year
```

```python
covid_yearly.to_csv('covid_year_cleaned.csv', index = False)
covid_year_month.to_csv('covid_year_month_cleaned.csv', index = False)
```
