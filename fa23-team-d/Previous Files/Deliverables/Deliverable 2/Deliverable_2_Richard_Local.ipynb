1) What type of building permits are approved each year by type (worktype), description, valuation (declared valuation), square footage, occupancy type?

```python
abp_base = abp[['issued_date','worktype', 'description', 'declared_valuation', 'sq_feet', 'occupancytype']]
abp_base['issued_date'] = pd.to_datetime(abp_base['issued_date'])
abp_base['issued_year']= abp_base['issued_date'].dt.year
abp_base = abp_base[['issued_year','worktype', 'description', 'declared_valuation', 'sq_feet', 'occupancytype']]
result_worktype = abp_base.groupby('issued_year')['worktype'].agg(lambda x: x.value_counts().to_dict())
worktype_result = result_worktype.reset_index()
expanded_categories = pd.DataFrame.from_records(worktype_result['worktype'])
worktype_result = pd.concat([worktype_result, expanded_categories], axis=1)
worktype_result = worktype_result.drop('worktype', axis=1)
worktype_result.set_index('issued_year', inplace=True)
worktype_result
# Extracting the column names representing building types
building_types = worktype_result.columns[1:]  # Assuming the first column is for years

# Summing up the buildings for each type across all years
building_totals = worktype_result[building_types].sum()

# Sorting the building types based on their total count
top_buildings = building_totals.nlargest(5)

# Getting the top 5 building types
top_building_types = top_buildings.index.tolist()

# Plotting the top 5 building types over the years
plt.figure(figsize=(10, 6))

for building_type in top_building_types:
    plt.plot(worktype_result.index, worktype_result[building_type], label=building_type)

plt.xlabel('Year')
plt.ylabel('Number of Buildings')
plt.title('Top 5 Building Types Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
result_description = abp_base.groupby('issued_year')['description'].agg(lambda x: x.value_counts().to_dict())
description_result = result_description.reset_index()
expanded_description = pd.DataFrame.from_records(description_result['description'])
description_result = pd.concat([description_result, expanded_description], axis=1)
description_result = description_result.drop('description', axis=1)
description_result.set_index('issued_year', inplace=True)
description_result

# Extracting the column names representing building types
description_types = description_result.columns[1:]  # Assuming the first column is for years

# Summing up the buildings for each type across all years
description_totals = description_result[description_types].sum()

# Sorting the building types based on their total count
top_descriptions = description_totals.nlargest(5)

# Getting the top 5 building types
top_description_types = top_descriptions.index.tolist()

# Plotting the top 5 building types over the years
plt.figure(figsize=(10, 6))

for description_type in top_description_types:
    plt.plot(description_result.index, description_result[description_type], label=description_type)

plt.xlabel('Year')
plt.ylabel('Number of Buildings')
plt.title('Top 5 Permitting by Description Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
result_occupancytype = abp_base.groupby('issued_year')['occupancytype'].agg(lambda x: x.value_counts().to_dict())
occupancytype_result = result_occupancytype.reset_index()
expanded_occupancytype = pd.DataFrame.from_records(occupancytype_result['occupancytype'])
occupancytype_result = pd.concat([occupancytype_result, expanded_occupancytype], axis=1)
occupancytype_result = occupancytype_result.drop('occupancytype', axis=1)
occupancytype_result.set_index('issued_year', inplace=True)
# Extracting the column names representing building types
occupancy_types = occupancytype_result.columns[1:]  # Assuming the first column is for years

# Summing up the buildings for each type across all years
occupancy_totals = occupancytype_result[occupancy_types].sum()

# Sorting the building types based on their total count
top_occupancies = occupancy_totals.nlargest(5)

# Getting the top 5 building types
top_occupancy_types = top_occupancies.index.tolist()

# Plotting the top 5 building types over the years
plt.figure(figsize=(10, 6))

for occupancy_type in top_occupancy_types:
    plt.plot(occupancytype_result.index, occupancytype_result[occupancy_type], label=occupancy_type)

plt.xlabel('Year')
plt.ylabel('Number of Buildings')
plt.title('Top 5 Permitting by Occupancy Types Over the Years')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Define the bounds for categorizing square footage
bounds = [0, 5000, 10000, 15000, 20000, float('inf')]  # Modify the bounds as needed

# Create labels for the square footage categories
labels = [f"{bounds[i]}-{bounds[i+1]}" if i != len(bounds) - 2 else f"{bounds[i]}+" for i in range(len(bounds)-1)]

# Categorize the square footage into the defined ranges
abp_base['declared_valuation_category'] = pd.cut(abp_base['declared_valuation'], bins=bounds, labels=labels, include_lowest=True)

result_declared_valuation = abp_base.groupby('issued_year')['declared_valuation_category'].agg(lambda x: x.value_counts().to_dict())
declared_valuation_result = result_declared_valuation.reset_index()
expanded_declared_valuation = pd.DataFrame.from_records(declared_valuation_result['declared_valuation_category'])
declared_valuation_result = pd.concat([declared_valuation_result, expanded_declared_valuation], axis=1)
declared_valuation_result = declared_valuation_result.drop('declared_valuation_category', axis=1)
declared_valuation_result.set_index('issued_year', inplace=True)

plt.figure(figsize=(10, 6))
for column in declared_valuation_result.columns:
    plt.plot(declared_valuation_result.index, declared_valuation_result[column], label=column)

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Line Chart of Declared Valuation by bounds over time')
plt.legend()
plt.grid(True)

plt.show()
```

```python
# Define the bounds for categorizing square footage
bounds_sqf = [0, 10, 20, 30, 40, float('inf')]  # Modify the bounds as needed

# Create labels for the square footage categories
labels = [f"{bounds_sqf[i]}-{bounds_sqf[i+1]}" if i != len(bounds_sqf) - 2 else f"{bounds_sqf[i]}+" for i in range(len(bounds_sqf)-1)]

# Categorize the square footage into the defined ranges
abp_base['sq_footage_category'] = pd.cut(abp_base['sq_feet'], bins=bounds_sqf, labels=labels, include_lowest=True)

result_sqft = abp_base.groupby('issued_year')['sq_footage_category'].agg(lambda x: x.value_counts().to_dict())
sqft_result = result_sqft.reset_index()
expanded_sqft = pd.DataFrame.from_records(sqft_result['sq_footage_category'])
sqft_result = pd.concat([sqft_result, expanded_sqft], axis=1)
sqft_result = sqft_result.drop('sq_footage_category', axis=1)
sqft_result.set_index('issued_year', inplace=True)

plt.figure(figsize=(10, 6))
for column in sqft_result.columns:
    plt.plot(sqft_result.index, sqft_result[column], label=column)

plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Line Chart of Square Footage by bounds over time')
plt.legend()
plt.grid(True)

plt.show()
```

What are the year over year trends visible in the zoning board of appeal approvals and denials by geography (neighborhood - listed as city, zip code, zoning district)? You’ll want to normalize the data, perhaps ratio of permits to approvals or denials, etc.

```python
zba_decision = zba[['final_decision_date', 'decision', 'city', 'zip', 'zoning_district']]
zba_decision
```

```python
zba_decision['final_decision_date'] = pd.to_datetime(zba_decision['final_decision_date'])
zba_decision['final_decision_year']= zba_decision['final_decision_date'].dt.year
```

```python
zba_decision = zba_decision[['final_decision_year', 'city', 'zip', 'zoning_district', 'decision']]
```

```python
zba_decision
```

```python
grouped_decision = zba_decision.groupby(['final_decision_year', 'zoning_district', 'decision']).size().unstack(fill_value=0)

# Calculate the ratio of approved to denied projects for each city in each year
grouped_decision['Ratio_Approved_to_Denied'] = grouped_decision['Approved'] / grouped_decision['Denied']
# Replace NaN values resulting from division by zero with 0 or any other desired value
grouped_decision['Ratio_Approved_to_Denied'].fillna(0, inplace=True)

# Resetting the index to have a flat structure
grouped_decision = grouped_decision.reset_index()

# View the resulting DataFrame with ratios
print(grouped_decision[['final_decision_year', 'zoning_district', 'Ratio_Approved_to_Denied']])
```

```python
# Filter out any rows where the ratio is 0 (optional)
grouped_decision_filtered = grouped_decision[grouped_decision['Ratio_Approved_to_Denied'] != 0]

# Plotting the ratio of approved to denied projects for each city in each year
plt.figure(figsize=(10, 6))

for district in grouped_decision_filtered['zoning_district'].unique():
    city_data = grouped_decision_filtered[grouped_decision_filtered['zoning_district'] == district]
    plt.plot(city_data['final_decision_year'], city_data['Ratio_Approved_to_Denied'], label= district)

plt.xlabel('Year')
plt.ylabel('Ratio of Approved to Denied Projects')
plt.title('Ratio of Approved to Denied Projects by City Over Years')
plt.legend(title='City', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
# Assuming 'grouped_data' contains the DataFrame with the ratios of approved to denied projects

# Filter out any rows where the ratio is 0 (optional)
grouped_decision_filtered = grouped_decision[grouped_decision['Ratio_Approved_to_Denied'] != 0]

# Remove or replace infinity or extremely large values with NaN
grouped_decision_filtered['Ratio_Approved_to_Denied'].replace([np.inf, -np.inf], np.nan, inplace=True)
grouped_decision_filtered['Ratio_Approved_to_Denied'].fillna(0, inplace=True)  # Replace NaN values with 0
# Normalize the ratio using Min-Max scaling
scaler = MinMaxScaler()
grouped_decision_filtered['Normalized_Ratio'] = scaler.fit_transform(grouped_decision_filtered[['Ratio_Approved_to_Denied']])

# Select a subset of districts (e.g., top 10 districts by the count of projects)
top_districts = grouped_decision_filtered.groupby('zoning_district')['Normalized_Ratio'].count().nlargest(10).index

# Filter data for top districts
top_districts_data = grouped_decision_filtered[grouped_decision_filtered['zoning_district'].isin(top_districts)]

# Create a FacetGrid of line plots for each district showing the normalized ratio
g = sns.FacetGrid(top_districts_data, col='zoning_district', col_wrap=3, height=4, aspect=1.5, sharey=False)
g.map(plt.plot, 'final_decision_year', 'Normalized_Ratio')

# Set titles for each subplot
g.set_titles("{col_name}")

# Set common labels and title
g.set_axis_labels('Year', 'Normalized Ratio of Approved to Denied Projects')
g.fig.suptitle('Normalized Ratio of Approved to Denied Projects for Top Districts Over Years', y=1.02)
plt.tight_layout()
plt.show()
```
