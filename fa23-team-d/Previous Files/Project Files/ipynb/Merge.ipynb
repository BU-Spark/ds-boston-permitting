```python
abp_df = pd.read_csv("/content/drive/MyDrive/City of Boston: Permitting D/Project Files/data/cleaned_abp.csv")
abp_df.head()
```

```python
zba_df = pd.read_csv('/content/drive/MyDrive/City of Boston: Permitting D/Project Files/data/cleaned_zba.csv', encoding='utf-8')
zba_df.head()
```

```python
abp_df.info() # object_id, city, zip, description
```

```python
zba_df.info() # boa_apno, city, zip, project_description
```

```python
abp_df['Approved'] = 1
zba_df['Approved'] = 0
```

```python
pre_abp_df = abp_df[['object_id','city', 'zip', 'description', 'Approved']]
pre_zba_df = zba_df[['boa_apno', 'city', 'zip', 'project_description', 'Approved']]
```

```python
pre_abp_df.rename(columns={'object_id': 'ID', 'description': 'Description'}, inplace=True)
pre_zba_df.rename(columns={'boa_apno': 'ID', 'project_description': 'Description'}, inplace=True)
```

```python
combined_df = pd.concat([pre_abp_df, pre_zba_df], ignore_index=True)
```
