## zip

```python
df.zip.head()
```

```python
df.zip.unique()
```

```python
def clean_zip(value):
    # If value is NaN, return it as is
    if pd.isna(value):
        return value

    value = str(value)

    # Remove hyphen and following characters
    if '-' in value:
        value = value.split('-')[0]

    # Remove '.0' from strings that were originally float values
    if value.endswith('.0'):
        value = value[:-2]

    # Prepend '0' to 4-digit numbers
    if len(value) == 4:
        value = '0' + value

    # Convert standalone small numbers to NaN
    if value.isdigit() and len(value) <= 3:
        return np.nan

    return value

# Apply the cleaning function to each element in the data list
df['zip'] = [clean_zip(z) for z in df['zip']]
```
